```python
 def population(top,bottom,shoes,neck,handbag,n):
        
        population2= pd.DataFrame(columns=['TOP', 'BOTTOM','SHOES','NECK', 'HANDBAG'])
        for i in range(1,n):
            t1=top.sample()
            b1=bottom.sample()
            s1=shoes.sample()
            n1=neck.sample()
            h1=handbag.sample()
            new_row = {'TOP':t1['ID'].values, 'BOTTOM':b1['ID'].values, 'SHOES':s1['ID'].values, 'NECK':n1['ID'].values, 'HANDBAG':h1['ID'].values}
            population2 = population2.append(new_row, ignore_index=True)
        return population2
    
```

```python
def color(data,population1,kind,inputt):  
    var=0
    array=[]
    for index, row in population1.iterrows():
        x=int(row[kind])
        t=data.loc[data['ID'] == x].COLOR.values
        if t==inputt:
            var=1
        if t!=inputt:
            var=0    
        array.append(var)
    return array
        

def code (data,population1,kind,inputt):
    var=0
    array=[]
    for index, row in population1.iterrows():
        x=int(row[kind])
        t=data.loc[data['ID'] == x].DRESS_CODE.values
        if t==inputt:
            var=1
        if t!=inputt:
            var=0
        array.append(var)
    return array
```

```python
def fit(population1,color1,code1,budget1,top,bottom,shoes,neck,handbag):
    
    topcolor=color(top,population1,'TOP',color1)
    botoomcolor=color(bottom,population1,'BOTTOM',color1)
    shoescolor=color(shoes,population1,'SHOES',color1)
    neckcolor=color(neck,population1,'NECK',color1)
    handbagcolor=color(handbag,population1,'HANDBAG',color1)
    
    topcode=code(top,population1,'TOP',code1)
    botoomcode=code(bottom,population1,'BOTTOM',code1)
    shoescode=code(shoes,population1,'SHOES',code1)
    neckcode=code(neck,population1,'NECK',code1)
    handbagcode=code(handbag,population1,'HANDBAG',code1)
    
    # budget fitness
    array = []
    for index, row in population1.iterrows():
        top_id = int(row['TOP'])-1
        bottom_id = int(row['BOTTOM'])-1
        shoes_id = int(row['SHOES'])-1
        neck_id = int(row['NECK'])-1
        handbag_id = int(row['HANDBAG'])-1
        
        cost = top.loc[top_id, 'PRICE'] + \
               bottom.loc[bottom_id, 'PRICE'] + \
               shoes.loc[shoes_id, 'PRICE'] + \
               neck.loc[neck_id, 'PRICE'] + \
               handbag.loc[handbag_id, 'PRICE']
        var = 1.0
        if cost > budget1:
            var = 0.0
        array.append(var)
    
    fit=[]
    for i in range(0,len(population1)):
        fit.append((0.2*((topcolor[i]+botoomcolor[i]+shoescolor[i]+neckcolor[i]+handbagcolor[i])/5))\
                   +(0.4*((topcode[i]+botoomcode[i]+shoescode[i]+neckcode[i]+handbagcode[i])/5))\
                  +(0.4*(array[i])))
    return fit
```

```python
def roulette_wheel(population):
    selected = pd.DataFrame(columns=['TOP', 'BOTTOM','SHOES','NECK', 'HANDBAG','FITNESS'])
    sum1= sum(population['FITNESS'])
    r=random.uniform(0,sum1)
    sum2=0
    for index, row in population.iterrows():
        sum2=sum2+row['FITNESS']
        if sum2>r:
            selected = selected.append(row, ignore_index=True)
            return selected
    return selected
```

```python
PIECE_LABELS = [('TOP', top), ('BOTTOM', bottom), ('SHOES', shoes), ('NECK', neck), ('HANDBAG', handbag)]

def show_piece_solution(sol, name, p_name):
    p_id = int(sol[name])-1
    print(name + ':', list(p_name.loc[p_id])[1:-1])

def show_solution(sol):
    for name, p_name in PIECE_LABELS:
        show_piece_solution(sol, name, p_name)
    print('FITNESS:', float(sol['FITNESS']))
```

```python
LABELS = ['TOP', 'BOTTOM', 'SHOES', 'NECK', 'HANDBAG']
N = len(LABELS)

def crossover(parent1, parent2, color, code, budget, crossover_rate = 1.0):
    u = random.random()
    
    # If the dataframes are not copied then we can have some unexpected results
    # We must make copies
    
    child1 = parent1.copy()
    child2 = parent2.copy()
    
    
    if u <= crossover_rate:
        point = random.randrange(0, N-1)
        for i in range(0, N):
            label = LABELS[i]
            
            if i < point:
                child1[label] = parent2[label]
                child2[label] = parent1[label]
            else:
                child1[label] = parent1[label]
                child2[label] = parent2[label]
    
    # We must recalculate fitness
    child1['FITNESS'] = fit(child1,color,code,budget,top,bottom,shoes,neck,handbag)
    child2['FITNESS'] = fit(child2,color,code,budget,top,bottom,shoes,neck,handbag)
        
    return child1, child2
    
```

```python
LABELS = ['TOP', 'BOTTOM', 'SHOES', 'NECK', 'HANDBAG']
PIECES = [top, bottom, shoes, neck, handbag]
N = len(PIECES)

def mutate(unit, color, code, budget, mutation_rate = 0.01):
    result = unit.copy()
    
    # if u is higher than mutation rate then we do not apply mutation
    u = random.random()
    if u >= mutation_rate:
        return result
    
    # mutation is done by replacing random piece of outfit
    # with another piece of outfit
    point = random.randrange(0, N)
    piece = PIECES[point]
    label = LABELS[point]
    new_sample = piece.sample()
    
    result[label] = [new_sample['ID'].values]
    
    # we must recalculate fitness
    result['FITNESS'] = fit(child,color,code,budget,top,bottom,shoes,neck,handbag)
    return result
    
```

```python
MAX_ITER = 25

def best_solution(pop):
    best = None
    for index, row in pop.iterrows():
        if best is None or int(row['FITNESS']) > int(best['FITNESS']):
            best = row
    return best

def pop_add(pop, p):
    new_row = {'TOP':p['TOP'], 'BOTTOM':p['BOTTOM'], 'SHOES':p['SHOES'], 'NECK':p['NECK'], 'HANDBAG':p['HANDBAG']}
    return pop.append(new_row, ignore_index=True)

def GA(color, code, budget, population_size=21, mutation_rate=0.01, crossover_rate=0.9,repeat=1):
    graph_data = pd.DataFrame(columns=['ITERATION', 'FITNESS'])
    
    for r in range(0, repeat):
        print('repeat:', r)
        # intial population
        pop = population(top,bottom,shoes,neck,handbag,population_size)

        # calculate fitness for population
        pop_fit = fit(pop,color,code,budget,top,bottom,shoes,neck,handbag)
        pop['FITNESS'] = pop_fit

        for i in range(0, MAX_ITER):
            new_pop = pd.DataFrame(columns=['TOP', 'BOTTOM','SHOES','NECK', 'HANDBAG', 'FITNESS'])
            for j in range(0, population_size//2):
                parent1 = roulette_wheel(pop)
                parent2 = roulette_wheel(pop)

                child1, child2 = crossover(parent1, parent2,color,code,budget)
                child1 = mutate(child1,color,code,budget)
                child2 = mutate(child2,color,code,budget)

                new_pop = pop_add(new_pop, child1)
                new_pop = pop_add(new_pop, child2)

            new_pop['FITNESS'] = fit(new_pop,color,code,budget,top,bottom,shoes,neck,handbag)
            pop = new_pop
            best_solution_in_iteration = best_solution(new_pop)
        
            graph_data_new_row = {'ITERATION':i, 'FITNESS':best_solution_in_iteration['FITNESS']}
            
            if r == 0:
                graph_data = graph_data.append(graph_data_new_row, ignore_index=True)
            else:
                graph_data.loc[i, 'FITNESS'] = graph_data.loc[i, 'FITNESS'] + graph_data_new_row['FITNESS']

            if repeat == 1 and i % 10 == 0:
                print('Iteration:', i, 'fitness:', best_solution_in_iteration['FITNESS'])

        if repeat == 1:
            solution = best_solution(pop)
            show_solution(solution)
    
    graph_data['FITNESS'] = graph_data['FITNESS']/repeat
    return graph_data
    
```

```python
mutation_rates = [0.001, 0.03, 0.5]
crossover_rates = [0.05, 0.9, 1.0]
population_sizes = [10, 20, 30]

parameters = [(m,c,p) for m in mutation_rates for c in crossover_rates for p in population_sizes]
parameters
```

```python
MAX_ITER = 25

index = 0

fitness_by_mcp = pd.DataFrame(columns=['MUTATION', 'CROSSOVER', 'POPULATION', 'FITNESS'])

for m, c, p in parameters:
    graph_data = GA(color1, code1, budget1, p, c, m, repeat=20)
    graph_data.plot(y='FITNESS', x='ITERATION', title='Fitness with parameters(population_size='+str(p)+',crossover_rate='+str(c)+',mutation_rate='+str(m)+')', figsize=(10,6), ylim=(0, 1))
    plt.savefig('plot{}.png'.format(index))
    index += 1
    plt.show()
    
    new_row = {'MUTATION': m, 'CROSSOVER': c, 'POPULATION': p, 'FITNESS': graph_data.loc[MAX_ITER-1, 'FITNESS']}
    fitness_by_mcp = fitness_by_mcp.append(new_row, ignore_index=True)
```
